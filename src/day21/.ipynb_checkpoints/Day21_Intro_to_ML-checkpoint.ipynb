### Task 1: The Paradigm Shift (Traditional vs. Machine Learning)
#### A short explanation of Machine Learning in your own words.


* Machine Learning is a part of Artificial Intelligence that helps computers learn from data and improve automatically without being programmed step by step for every task.
* Machine Learning enables computers to learn from experience and make smarter decisions over time.
* For example, if we show many emails marked as spam and not spam, the system learns the difference and can identify new spam emails on its own.
