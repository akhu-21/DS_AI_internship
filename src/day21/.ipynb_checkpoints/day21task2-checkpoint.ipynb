# Day21 – Types of Machine Learning

## 1. Supervised Learning

Supervised learning uses data that contains inputs and correct output labels.
The model learns the relationship between input and output by reducing prediction errors.
It is mainly used for:

Classification → predicting categories

Regression → predicting numerical values

Example: Predicting house price, spam email detection.

---

## 2. Unsupervised learning uses data without labels.
* The model explores the data and finds patterns, groups, or structures on its own.
* Unsupervised Learning is a type of machine learning where the dataset does not contain labels.

Example: Grouping similar customers, identifying hidden trends.
---

## 3. Reinforcement Learning

Reinforcement learning works through interaction and feedback.
An agent takes actions in an environment and receives rewards or penalties.
The goal is to learn actions that give the maximum total reward.

Example: Game-playing AI, self-driving decision systems.

---

## Comparison Table

| Learning Type              | Real-World Example     | What the Model Learns                            | Data Type       |
| -------------------------- | ---------------------- | ------------------------------------------------ | --------------- |
| **Supervised Learning**    | Spam Email Detection   | Learns to classify emails using labeled examples | Labeled data    |
|                            | House Price Prediction | Learns relationship between features and price   | Labeled data    |
| **Unsupervised Learning**  | Customer Segmentation  | Groups similar customers based on behavior       | Unlabeled data  |
|                            | Market Basket Analysis | Finds products often bought together             | Unlabeled data  |
| **Reinforcement Learning** | Game Playing AI        | Learns best moves through rewards and penalties  | Reward feedback |
|                            | Self-Driving Decisions | Learns driving actions that maximize safety      | Reward feedback |


---

## Why Choosing the Correct Type Matters

Choosing the wrong learning type can make your model fail before development begins. 
If you use supervised learning without labeled data, the model cannot learn correctly. 
If you use unsupervised learning when prediction is required, it will not provide accurate outputs. 
Reinforcement learning is suitable only when learning happens through interaction and rewards.
