# Phase 1: The Detective Work (Setup & Inspection)

```python
import pandas as pd 
df=pd.read_csv('customer_analytics.csv')
df.head()
df.info()
df.describe()
```

1. This dataset contains customer analytics information used to understand customer behavior and purchasing patterns.Each row represents one customer       and their demographic details, income level, shopping activity, and engagement with the company.
2. The head() output displays the first few records, helping us understand the structure and sample values of the dataset.
3. The info() output shows data types and confirms the presence of missing values,
4. describe() provides summary statistics like mean, min, max, and standard deviation for numerical columns.

# Phase 2: The Cleanup (Data Preprocessing)

```python
df.isnull().sum()
```

```python
df["Education"]=df["Education"].fillna(df["Education"].mode()[0])
df["AnnualIncome"] = df["AnnualIncome"].fillna(df["AnnualIncome"].median())
df.isnull().sum()
```

1.Missing values in categorical and numerical columns were handled using mode and median respectively.
2.The output confirms that all missing values have been successfully treated.

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

 1.This output checks the number of duplicate rows in the dataset.It ensures that repeated records do not bias the analysis.
 
 2. The second duplicate check confirms whether duplicate entries exist after preprocessing.A value of zero indicates the dataset contains only unique records.

 # Phase 3: The Deep Dive (Univariate & Bivariate Analysis)

```python
import matplotlib.pyplot as plt
import seaborn as sns
#univarient
sns.set(style="whitegrid")
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=10)
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

1. Most customers fall in the middle-age range. The data shows a balanced spread with no extreme outliers.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(6,4))
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()#univarient
```

The chart shows the proportion of male and female customers. One category(male) slightly dominates the dataset.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(6,4))
sns.histplot(df["AnnualIncome"], bins=10)
plt.title("Annual Income Distribution")
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()
```

Customer income varies widely, indicating different purchasing capacities across the dataset.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(6,4))
plt.scatter(df["AnnualIncome"], df["SpendingScore"])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Income vs Spending Behavior")
plt.show()
```

Customers with higher income do not always spend more. Spending behavior varies across income levels.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(7,4))
sns.boxplot(x="Education", y="LastPurchaseAmount", data=df)
plt.title("Purchase Amount by Education Level")
plt.xticks(rotation=30)
plt.show()
```

Purchase amounts differ across education groups, showing variation in spending patterns.

```python
plt.figure()
plt.boxplot(df['AnnualIncome'].dropna())
plt.title("AnnualIncome")
plt.show()
```

The box plot summarizes income distribution and identifies extreme values (outliers).
It provides a clear view of median income and overall spread.

```python
plt.figure()
sns.boxplot(x='Gender', y='SpendingScore', data=df)
plt.title("Spending Score by Gender")
plt.show()
```

The box plot compares spending behavior across genders.
It highlights differences in median spending score and shows any outliers.

# Phase 4: The Big Picture (Multivariate & Storytelling)

```python
corr = df.corr(numeric_only=True)
corr
```

The correlation matrix shows the strength and direction of relationships between numerical variables.
Values closer to +1 or -1 indicate strong relationships, while values near 0 indicate weak relationships.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

The heatmap visually represents correlations between multiple variables using color intensity.
It helps quickly identify strong positive or negative relationships in the dataset.
