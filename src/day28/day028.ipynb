```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\madso\DS_AI_Internship\src\day28\insurance_data.csv")
df.head()
```

```python
df.shape
```

```python
plt.scatter(df.age,df.bought_insurance,color = 'red',marker='*' )
```

```python
x = df[['age']]
y = df[['bought_insurance']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted
```

```python
df['probs'] = model.predict_proba(df[['age']])[:,1]
```

```python
df
```

```python
plt.scatter(df['age'],df['bought_insurance'],color = 'red',marker = '*')
plt.plot(df['age'],df['probs'],color = 'blue')
```

```python
import matplotlib.pyplot as plt
df_sorted = df.sort_values(by='age')
plt.scatter(df_sorted['age'], df_sorted['bought_insurance'], color='red', marker='*')
plt.plot(df_sorted['age'], df_sorted['probs'], color='blue')
plt.xlabel("Age")
plt.ylabel("Probability")
plt.title("Sigmoid Curve")
plt.show()
```
